```python
import os
import shutil
import pandas as pd
```

```python
df = pd.read_csv("pairs_26.csv")
# print(df)
```

```python
# Function to extract folder names from the pandas DataFrame
def extract_folder_names(df):
    folder_names = []
    for pair in df['column1']:
        folder_names.extend(pair.split('::')[0].split(':'))
        folder_names.extend(pair.split('::')[1].split(':'))
    return folder_names
```

```python
def copy_folders(dataframe, source_folder, destination_folder):
    # Create destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Extract folder names from the dataset
    folder_names = extract_folder_names(dataframe)
    print("folder len:",len(folder_names))

    # Filter folder names to include only unique ones
    # folder_names = list(set(folder_names))
    # print("no. of unique folders:",len(folder_names))

    # Copy folders
    for folder_name in folder_names:
        source_path = os.path.join(source_folder, folder_name)
        destination_path = os.path.join(destination_folder, folder_name)
        try:
            shutil.copytree(source_path, destination_path)
            print(f"Folder '{folder_name}' copied successfully.")
        except FileNotFoundError:
            print(f"Folder '{folder_name}' not found.")
        except FileExistsError:
            print(f"Folder '{folder_name}' already exists in the destination.")
```

```python
# Paths and filenames
# dataset_file = 'pairs_26.txt'
source_folder = 'images'
destination_folder = 'new_images_rest_26'

# extract_folder_names(df)
copy_folders(df,source_folder,destination_folder)
```

```python
# Rename images
import os

def rename_images_in_folder(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter only image files
    image_files = [file for file in files if os.path.isfile(os.path.join(folder_path, file))]
    
    # Rename image files
    for i, image_file in enumerate(image_files):
        # Get file extension
        _, ext = os.path.splitext(image_file)
        # New filename
        new_filename = f"image_{i+100+1}{ext}"
        # Rename the file
        os.rename(os.path.join(folder_path, image_file), os.path.join(folder_path, new_filename))
        print(f"Renamed: {image_file} to {new_filename}")

def rename_images_in_all_subfolders(main_folder):
    # Iterate over all subfolders
    for folder_name in os.listdir(main_folder):
        folder_path = os.path.join(main_folder, folder_name)
        if os.path.isdir(folder_path):
            print(f"Processing folder: {folder_name}")
            rename_images_in_folder(folder_path)

# Replace 'images' with the path to your main images folder
main_folder = 'new_images'

# Call the function to rename images in all subfolders
rename_images_in_all_subfolders(main_folder)
```
